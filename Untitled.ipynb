```python
import os
import pandas as pd
import matplotlib.pyplot as plt
cwd = os.getcwd()
```

```python
# import data
df = pd.read_excel(cwd + "/Results/angSep_CASCOT__2q_EE.xls", header=None, names=['date', 'angSep_2q_EE'])
df.date=pd.to_datetime(df.date)
df.set_index(['date'],inplace=True)

temp_df=pd.read_excel(cwd + "/Results/angSep_CASCOT__5q_IEUE.xls", header=None, names=['date', 'angSep_5q_IE_UE'])
temp_df.date=pd.to_datetime(temp_df.date)
temp_df.set_index(['date'],inplace=True)

df_new = df.join(temp_df)

temp_df=pd.read_excel(cwd + "/Results/modOfMod_CASCOT__2q_EE.xls", header=None, names=['date', 'modOfMod_2q_EE'])
temp_df.date=pd.to_datetime(temp_df.date)
temp_df.set_index(['date'],inplace=True)

df_new = df_new.join(temp_df)

temp_df=pd.read_excel(cwd + "/Results/modOfMod_CASCOT__5q_IEUE.xls", header=None, names=['date', 'modOfMod_5q_IE_UE'])
temp_df.date=pd.to_datetime(temp_df.date)
temp_df.set_index(['date'],inplace=True)

df_new = df_new.join(temp_df)


temp_df=pd.read_excel(cwd + "/Results/EE_2q.xls", header=None, names=['date', 'EE_2q'])
temp_df.date=pd.to_datetime(temp_df.date)
temp_df.set_index(['date'],inplace=True)

df_new = df_new.join(temp_df)

temp_df=pd.read_excel(cwd + "/Results/IEorUE_5q.xls", header=None, names=['date', 'IEorUE_5q'])
temp_df.date=pd.to_datetime(temp_df.date)
temp_df.set_index(['date'],inplace=True)

df_new = df_new.join(temp_df)

temp_df=pd.read_excel(cwd + "/Results/Aggre_Ur_pct_2q.xls", header=None, names=['date', 'Aggre_Ur_pct_2q'])
temp_df.date=pd.to_datetime(temp_df.date)
temp_df.set_index(['date'],inplace=True)

df_new = df_new.join(temp_df)

df_new
```

```python

df_new.angSep_2q_EE = df_new.angSep_2q_EE.rolling(5, center=True).mean()
df_new.angSep_5q_IE_UE = df_new.angSep_5q_IE_UE.rolling(5, center=True).mean()
df_new.modOfMod_2q_EE = df_new.modOfMod_2q_EE.rolling(5, center=True).mean()
df_new.modOfMod_5q_IE_UE = df_new.modOfMod_5q_IE_UE.rolling(5, center=True).mean()
df_new.EE_2q = df_new.EE_2q.rolling(5, center=True).mean()/10000000000
df_new.IEorUE_5q = df_new.IEorUE_5q.rolling(5, center=True).mean()/10000000000
#df_new.Aggre_Ur_pct_2q = df_new.Aggre_Ur_pct_2q.rolling(5, center=True).mean()
```
